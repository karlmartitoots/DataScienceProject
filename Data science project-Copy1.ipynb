# Weather data

```python
w_d_web = 'http://meteo.physic.ut.ee/et/archive.php?do=data&begin%5Byear%5D=2019&begin%5Bmon%5D=11&begin%5Bmday%5D=4&end%5Byear%5D=2019&end%5Bmon%5D=12&end%5Bmday%5D=5&9=1&16=1&14=1&ok=+Esita+p%C3%A4ring+'

table = requests.get(w_d_web)
```

```python
w_d_raw = []

for row in table:
    w_d_raw.append(row.decode('utf-8').split('\n')) # Removing all non-data characters (\n and b)
```

```python
w_d_raw
```

```python
string = ''
for row in w_d_raw:
    for el in row:
        for x in el:
            string += str(x) # Making of a continuous string for the separation of data
    
print(string)
```

```python
s_string = string.split(', ')
```

```python
date = []
time = []
l = 0

for row in s_string:
    if bool(re.search(':', row)) == True: # However, the data from web had some peculiarities - irradiation flux and
        t_d = (row[-19:]).split(' ')      # date/time were in the same "cell" for some reason, they had to be
        n = 0                             # separated for the dataframe
        
        while n < len(t_d):               # Separating date and time
            if n == 0:
                date.append(t_d[n])
            if n == 1:
                time.append(t_d[n])
            n += 1
            
        s_string[l] = s_string[l].replace(s_string[l][-19:], '')
                
    l += 1
```

```python
temp = []
prec = []
flux = []
```

```python
k = 3
m = 1

while k < len(s_string) - 1: # Classifying weather data correctly
    k += 1

    if m == 1:
        temp.append(round(float(s_string[k]), 3))
        m += 1
        continue

    if m == 2:
        prec.append(round(float(s_string[k]), 3))
        m += 1
        continue

    if m == 3:
        flux.append(round(float(s_string[k]), 3))
        m = 1
        continue
    
    
```

```python
w_d = [date, time, temp, prec, flux]

weather_df = pd.DataFrame(columns = ['date', 'time', 'temp', 'prec', 'flux'])
```

```python
n = 0
for row in weather_df.columns:
    weather_df[row] = w_d[n]
    n += 1
```

```python
x_c = np.zeros((8, 1), dtype = int)
y_c = np.zeros((1, 8), dtype = int)
c = np.zeros((8, 8), dtype = int)
```

```python
def interval(x_start, y_start, interval):

    k = 1
    while k <= len(x_c):
        x_c[k-1][0] = x_start + (interval*k)
        y_c[0][k-1] = y_start + (interval*k)
        k += 1
```

```python
def grid_loc(a, b, l1, l2): # l1 and l2 are to determain the coordinates position in the made grid
    n = 0
    l = [0, 0]
    while n < len(l1):
        if a < l1[n][0]: # a is the x-coordinate
            l[0] = n
            break

        n += 1

    z = 0
    
    while z < len(l1):
        if b < l2[0][z]: # b is the y-coordinate
            l[1] = z
            break

        z += 1
```
